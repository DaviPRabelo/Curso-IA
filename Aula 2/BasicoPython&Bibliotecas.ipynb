### 1.1 Listas

Uma lista é uma coleção ordenada e variável.

-> Ordenada : Cada elemento tem uma posição (índice)

-> Mutável : Conseguimos alterar, adicionar e remover itens

```python
## Criação, acesso e operações básicas

numeros = [10, 20, 30, 40]
nomes = ["Ana", "Bruno", "Carla"]

print(f"Lista de números: {numeros}")
print(f"Lista de nomes: {nomes}")

#Acessando por indices
print(f"Primeiro elemento da lista de números: {numeros[0]}")
print(f'Segundo elemento da lista de nomes: {nomes[1]}')

#Tamanho de uma lista
print(f'Tamanho da lista de números: {len(numeros)}')
print(f'Tamanho da lista de nomes: {len(nomes)}')

#Fatiamento (slicing)
print(f"Fatiamento da lista de números: {numeros[1:3]}") # Do indicie 1 ate o indice 2
```

```python
# Append, extend, insert, remove, pop, sort

numeros = [10, 20, 30]

numeros.append(40) #Adiciona no final
numeros.extend([50,60]) #Adiciona varios no final
numeros.insert(1, 15) #Insere na posição 1

print(f" Depois das operações: {numeros}")

numeros.remove(20) #Remove o valor 20
ultimo = numeros.pop() #Remove e retorna o ultimo indice

print(f" Depois das operações: {numeros}")
print(f" Último elemento removido: {ultimo}")

#Ordenação
numeros.sort()
print(f"Lista ordenada: {numeros}")

```

###1.2 Dicionários

Um dicionário guarda paraes chave -> valor

- Ideal quando voce quer buscar dados por 'nome' (chave)
- Muito usado para configurações, contagem, mapeamento, json,etc


```python
#Criação, acesso e atualização
aluno = {
    "nome": "Pedro",
    "idade": 23,
    "curso": "Engenharia de Software",
    "notas": [8.5, 7.0, 9.2]
}

print("Aluno:", aluno)
print("Nome:", aluno["nome"])
print("Notas:", aluno["notas"])

# atualizar/adicionar
aluno["idade"] = 24
aluno["cidade"] = "São José dos Campos"

print("Atualizado:", aluno)
```

```python
#Iteração e alguns métodos
aluno = {"nome": "Lívia", "idade": 22, "curso": "Engenharia"}

print("Chaves:", list(aluno.keys()))
print("Valores:", list(aluno.values()))
print("Itens (chave, valor):", list(aluno.items()))

for chave, valor in aluno.items():
    print(f"{chave} -> {valor}")
```

```python
#Exemplo de contagem
# Quantas vezes cada categoria aparece
categorias = ["A", "B", "A", "C", "B", "A", "D", "C", "A"]

contagem = {}
for cat in categorias:
    contagem[cat] = contagem.get(cat, 0) + 1

print(contagem)

```

### 1.3 Estruturas de repetição (for/ while)
Loops sao como um motor para tarefas repetitivas

-> For: Quando voce sabe o que vai percorrer(lista, range dataframe, etc)

-> While: Quando voce repete até uma condição parar de ser verdadeira

```python
#Percorrendo uma lista com for
for i in range(5):
    print("i =", i)

valores = [2, 4, 6, 8]
soma = 0
for v in valores:
    soma += v
print("Soma:", soma)
```

```python
# Percorrendo uma lista com whiel
contador = 0
while contador < 3:
    print("contador =", contador)
    contador += 1
```

### 1.4 Compreensão de listas (list comprehension)

É uma forma compacta de criar listas

**Quando usar :** transformar/filtrar listas serapadamente

```python
# quadrados de 0 a 9
quadrados = [x**2 for x in range(10)]
print(quadrados)

# filtrar pares
pares = [x for x in range(20) if x % 2 == 0]
print(pares)
```

### Numpy (np)

O principal objeto é um *ndarray* ( array-n-dimensional), que é mais rapido e prático para cálculos do que listas comuns

```python
#!pip install numpy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("NumPy:", np.__version__)
print("Pandas:", pd.__version__)
```

```python
#Criando arrays e operações básiacs
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])

print("a:", a)
print("b:", b)

# operações vetorizadas (sem loop explícito)
print("a + b:", a + b)
print("a * 2:", a * 2)
print("a^2:", a ** 2)

print("Média de a:", a.mean())
print("Soma de a:", a.sum())
```

```python
#Reshape, operações e matrizes
m = np.arange(1, 13)  # 1..12
m2 = m.reshape(3, 4)  # 3 linhas, 4 colunas

print("m2:\n", m2)

print("Linha 0:", m2[0])
print("Coluna 1:", m2[:, 1])

print("Soma por linha:", m2.sum(axis=1))
print("Soma por coluna:", m2.sum(axis=0))
```

### Pandas

Pandas é fundamental para trabalhar com dados tabulares

-> Serie: Uma coluna
-> DataFrame  : Uma tabela (varias colunas)



```python
df = pd.DataFrame({
    "nome": ["Ana", "Bruno", "Carla", "Diego"],
    "idade": [20, 25, 23, 28],
    "cidade": ["SP", "RJ", "MG", "SP"],
    "salario": [2500, 4000, 3200, 5000]
})

df
```

```python
#Selecionando colunas e filtros
print("Coluna idade:")
display(df["idade"])

print("\nLinhas com cidade == 'SP':")
display(df[df["cidade"] == "SP"])

print("\nLinhas com salario >= 3500:")
display(df[df["salario"] >= 3500])
```

```python
#Estatísticas descritivas rápidas

display(df.describe())

print("\nTipos das colunas:")
print(df.dtypes)

print("\nValores únicos em cidade:")
print(df["cidade"].unique())
```

```python
#Groupby : Agrupamento por categoria

media_salario_por_cidade = df.groupby("cidade")["salario"].mean().sort_values(ascending=False)
media_salario_por_cidade
```

**Lidando com valores ausentes(NaN)**

É sempre importante descobrir onde tem valores faltando
* Descobrir onde tem NaN(isna)
* Decidir: Remover ou preencher(dropna, fillna)


```python
df_nan = df.copy()
df_nan.loc[2, "salario"] = np.nan  # cria um NaN proposital

display(df_nan)

print("\nQuantidade de NaN por coluna:")
print(df_nan.isna().sum())

print("\nPreenchendo NaN com a média da coluna salario:")
df_nan["salario"] = df_nan["salario"].fillna(df_nan["salario"].mean())

display(df_nan)
```

### 2.3 Matplotlib

Matplotlib é a biblioteca clássica de gráficos em Python.

Gráficos comuns em EDA:
- Histograma (distribuição)

```python
## Barras : Media por categoria
plt.figure()
media_salario_por_cidade.plot(kind="bar")
plt.title("Média de salário por cidade")
plt.xlabel("Cidade")
plt.ylabel("Média do salário")
plt.show()
```

## Dataset real (para praticar)

Como este ambiente pode não ter internet, vamos usar um dataset real que vem “embutido” no scikit-learn: **Iris**.

- 150 amostras
- 4 medidas de pétala/sépala
- 3 espécies (classe)


```python
## Carregando o datset da iris
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
df_iris = iris.frame.copy()
df_iris.rename(columns={"target": "species_id"}, inplace=True)

# cria uma coluna com o nome da espécie
df_iris["species"] = df_iris["species_id"].map(dict(enumerate(iris.target_names)))

df_iris.head()
```

Eda passo a passo:

A sequencia abaixo funciona para a maioria dos datasets

```python
# Checar a estrutura do dataset
print("Shape:", df_iris.shape)
print("\nTipos:")
print(df_iris.dtypes)

display(df_iris.head())
```

```python
#Checando valores ausentes
df_iris.isna().sum()
```

```python
#Estatísticas descritivas(numericas)
df_iris.describe()
```

```python
#Entendendo categorias (espécies)
df_iris["species"].value_counts()
```

```python
#Distribuição (histogramas) das features

cols = ["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"]

for col in cols:
    plt.figure()
    plt.hist(df_iris[col], bins=15)
    plt.title(f"Distribuição: {col}")
    plt.xlabel(col)
    plt.ylabel("Frequência")
    plt.show()
```

```python
# Comparação por categoria (media por especie)
media_por_especie = df_iris.groupby("species")[cols].mean()
media_por_especie
```

```python
#Grafico de barra : media de petal lenght por espécie

plt.figure()
media_por_especie["petal length (cm)"].plot(kind="bar")
plt.title("Média de petal length (cm) por espécie")
plt.xlabel("Espécie")
plt.ylabel("Média (cm)")
plt.show()
```

```python
#Relações entre variaveis(scatter)
plt.figure()
for sp in df_iris["species"].unique():
    sub = df_iris[df_iris["species"] == sp]
    plt.scatter(sub["petal length (cm)"], sub["petal width (cm)"], label=sp)

plt.title("Petal length x Petal width (por espécie)")
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.legend()
plt.show()
```

```python
#Correlações (para variaveis numericas)
corr = df_iris[cols].corr()
corr
```

```python
#  Heatmap simples (sem seaborn, só com matplotlib)


plt.figure()
plt.imshow(corr.values)
plt.title("Matriz de correlação (Iris)")
plt.xticks(range(len(cols)), cols, rotation=45, ha="right")
plt.yticks(range(len(cols)), cols)
plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


# 2. Selecionar apenas as colunas numéricas (removendo o alvo)
# Ajuste o nome das colunas se necessário
X = dts.select_dtypes(include=[np.number]).drop(columns=['diagnosed_diabetes'])
y = dts['diagnosed_diabetes']

# 3. PADRONIZAÇÃO (Passo obrigatório para PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. APLICAR PCA
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)

print(f"Dados originais: {X_scaled.shape[1]} colunas")
print(f"Dados após PCA: {X_pca.shape[1]} colunas (mantendo 90% da informação)")

# 5. INTEGRAR COM MACHINE LEARNING (Exemplo com Random Forest)
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2)
clf = RandomForestClassifier()
clf.fit(X_train, y_train)

print(f"Acurácia do modelo com PCA: {clf.score(X_test, y_test):.2%}")
```
